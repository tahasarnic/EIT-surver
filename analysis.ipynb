```python
# Question 9
df_ques_9 = df.iloc[:, df.columns.str.startswith('9.')].assign(regional_ino_index = df['2023-22-21-19 Regional Innovation Index'])
df_ques_9 = df_ques_9.fillna(0)
```

## Programme 1

```python
# Program type counts
ques_9_program_type_1 = df_ques_9.groupby('regional_ino_index')[df_ques_9.columns[0]].value_counts().reset_index().pivot_table(values = 'count', index = 'regional_ino_index', columns = '9. Please enter information on your support programme(s).  >> Programme-1 >> Programme Type').fillna(0)
ques_9_program_type_1
```

```python
# Cohort size categoration
def cat_cohort_size(data):
    if data == 0:
        return '0'
    elif data >= 1 and data <= 5:
        return '1-5'
    elif data > 5 and data <= 10:
        return '6-10'
    elif data > 10 and data <= 15:
        return '11-15'
    elif data > 15 and data <= 20:
        return '16-20'
    elif data > 20 and data <= 50:
        return '21-50'
    elif data > 50 and data <= 100:
        return '51-100'
    elif data > 100:
        return '100+'
```

```python
# Cohort size manipulation
df_ques_9[df_ques_9.columns[2]] = df_ques_9[df_ques_9.columns[2]].replace(['-', '10-100', '10-15', '10-20', '100+', '1000+', '13  initiatives', '13 startups', '140-180 projects', '15-20', '20-50 individuals', '25-40 projects', '3 - 6', '5 + 5', '5-6', 'X', 'average 10', 'average 20', 'no'], ['0', '55', '12.5', '15', '101', '1001', '13', '13', '160', '17.5', '35', '32.5', '4.5', '10', '5.5', '0', '10', '20', '0']).astype(float)
df_ques_9[df_ques_9.columns[2]] = df_ques_9[df_ques_9.columns[2]].apply(cat_cohort_size)
df_ques_9[df_ques_9.columns[2]] = df_ques_9[df_ques_9.columns[2]].astype('category').cat.set_categories(['0', '1-5', '6-10', '11-15', '16-20', '21-50', '51-100', '100+'])
```

```python
# Program cohort size count
ques_9_cohort_1 = df_ques_9.groupby('regional_ino_index')[df_ques_9.columns[2]].value_counts().reset_index().pivot_table(values = 'count', index = 'regional_ino_index', columns = df_ques_9.columns[2]).fillna(0)
ques_9_cohort_1
```

```python
# Program stage focus count
ques_9_stage_1 = df_ques_9.groupby('regional_ino_index')[df_ques_9.columns[3]].value_counts().reset_index().pivot_table(values = 'count', index = 'regional_ino_index', columns = df_ques_9.columns[3]).fillna(0)
ques_9_stage_1
```

```python
# How many times did you implement the program manipulation
def cat_implement_times(data):
    try:
        if data == 0:
            return '0'
        elif data >= 1 and data <= 5:
            return '1-5'
        elif data > 5 and data <= 10:
            return '6-10'
        elif data > 10 and data <= 50:
            return '11-50'
        elif data > 50:
            return '50+'
    except:
        return 0
```

```python
# Average number of applicaitons manipulation
def cat_application_times(data):
    try:
        if data == 0:
            return '0'
        elif data >= 1 and data <= 10:
            return '1-10'
        elif data > 10 and data <= 20:
            return '11-20'
        elif data > 20 and data <= 40:
            return '21-40'
        elif data > 40 and data <= 100:
            return '41-100'
        elif data > 100 and data <= 300:
            return '101-300'
        elif data > 300:
            return '300+'
    except:
        return 0
```

```python
# Average number of applicaitons manipulation
df_ques_15[df_ques_15.columns[2]] = df_ques_15[df_ques_15.columns[2]].replace(['*', '-', 0, '1-2', '10-20', '20-40', '20-50', '3-6', '30-40', '?', 'na', 'not applicable'], ['0', '0', '0', '1.5', '15', '30', '35', '4.5', '35', '0', '0', '0']).astype('float')
df_ques_15[df_ques_15.columns[2]] = df_ques_15[df_ques_15.columns[2]].apply(cat_application_times)
df_ques_15[df_ques_15.columns[2]] = df_ques_15[df_ques_15.columns[2]].astype('category').cat.set_categories(['0', '1-10', '11-20', '21-40', '41-100', '101-300', '300+'])
```

```python
# Application counts
ques_15_application_1 = df_ques_15.groupby('regional_ino_index')[df_ques_15.columns[2]].value_counts().reset_index().pivot_table(values = 'count', index = 'regional_ino_index', columns = df_ques_15.columns[2]).fillna(0)
ques_15_application_1
```

```python
# Dropout percentage manipulation
df_ques_15[df_ques_15.columns[3]] = df_ques_15[df_ques_15.columns[3]].str.replace('%', '').str.strip()
df_ques_15[df_ques_15.columns[3]] = df_ques_15[df_ques_15.columns[3]].replace(['*', '-', '=', 'Dont know yet', 'We dont have that figure', 'less than 10', 'na', np.nan, 'nna', 'not applicable', '≈10', '≈15', '≈5'], ['0', '0', '0', '0', '0', '10', '0', '0', '0', '0', '10', '15', '5']).astype('float').fillna(0)
```
