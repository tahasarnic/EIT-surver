```python
#---Packages---
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
#---Paths---
ROOT_DIR = Path.cwd()
DATA_DIR = ROOT_DIR / 'data'
```

```python
# Data
df = pd.read_excel(DATA_DIR / 'EIT Survey Responses Total.xlsx')
```

# Question 9

```python
# 9. soru
df_ques_9 = df.iloc[:, df.columns.str.startswith('9.')].assign(regional_ino_index = df['2023-22-21-19 Regional Innovation Index'])
```

## Programme 1

```python
# Program type counts
df_ques_9.groupby('regional_ino_index')[df_ques_9.columns[0]].value_counts().reset_index().pivot_table(values = 'count', index = 'regional_ino_index', columns = '9. Please enter information on your support programme(s).  >> Programme-1 >> Programme Type').fillna(0)
```

```python
# Program duration counts
df_ques_9.groupby('regional_ino_index')[df_ques_9.columns[1]].value_counts().reset_index().pivot_table(values = 'count', index = 'regional_ino_index', columns = df_ques_9.columns[1]).fillna(0)
```

```python
# Cohort size manipulation
def cat_cohort_1(data):
    try:
        if data == 0:
            return '0'
        elif data >= 1 and data <= 5:
            return '1-5'
        elif data > 5 and data <= 10:
            return '6-10'
        elif data > 10 and data <= 15:
            return '11-15'
        elif data > 15 and data <= 20:
            return '16-20'
        else:
            return '20+'
    except:
        return np.nan
df_ques_9[df_ques_9.columns[2]] = df_ques_9[df_ques_9.columns[2]].replace(['15-20', 'average 20', '140-180 projects', '13 startups', '1000+', '3 - 6', '5 + 5', '10-100', '5-6'], ['17.5', '20', '160', '13', '1000', '4.5', '10', '55', '5.5']).astype('float')
df_ques_9[df_ques_9.columns[2]] = df_ques_9[df_ques_9.columns[2]].apply(cat_cohort_1)
```

```python
# Program cohort size count
df_ques_9.groupby('regional_ino_index')[df_ques_9.columns[2]].value_counts().reset_index().pivot_table(values = 'count', index = 'regional_ino_index', columns = df_ques_9.columns[2]).fillna(0)
```

```python
# Program stage focus count
df_ques_9.groupby('regional_ino_index')[df_ques_9.columns[3]].value_counts().reset_index().pivot_table(values = 'count', index = 'regional_ino_index', columns = df_ques_9.columns[3]).fillna(0)
```

```python
# Program type - duration vs. regional inovation index
df_ques_9.groupby(['regional_ino_index', df_ques_9.columns[0]])[df_ques_9.columns[1]].value_counts().reset_index().pivot(index = ['regional_ino_index', df_ques_9.columns[0]], columns = df_ques_9.columns[1], values = 'count').fillna(0)
```
